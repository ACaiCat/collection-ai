```python
import matplotlib.pyplot as plt

# 数据
frequency = [10, 100, 1000, 10000, 100000]  # 频率 (Hz)
equivalent_resistance = [41.67, 41.67, 86.96, 219.51, 4306.12]  # 等效电阻 (Ω)

# 绘制图表
plt.figure(figsize=(8, 2.5))
plt.plot(frequency, equivalent_resistance, marker='o', label="Equivalent Resistance")

# 设置对数坐标轴
plt.xscale('log')

# 添加标题和标签
plt.title("Equivalent Resistance vs Frequency", fontsize=14)
plt.xlabel("Frequency (Hz) - Log Scale", fontsize=12)
plt.ylabel("Equivalent Resistance (Ω)", fontsize=12)
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

#添加水印
plt.text(0.5, 0.9, 'By ZKS AND WJH', fontsize=20, color='gray', ha='center', va='center', alpha=0.3, transform=plt.gcf().transFigure)


# 显示图表
plt.show()
```

```python
# coding=utf-8
from matplotlib import pyplot as plt
import random
import matplotlib
from matplotlib import font_manager

#windws和linux设置字体的放
# font = {'family' : 'MicroSoft YaHei',
#         'weight': 'bold',
#         'size': 'larger'}   
# matplotlib.rc("font",**font)
# matplotlib.rc("font",family='MicroSoft YaHei',weight="bold")

#另外一种设置字体的方式
my_font = font_manager.FontProperties(fname="/System/Library/Fonts/PingFang.ttc")

x = range(0,120)
y = [random.randint(20,35) for i in range(120)]

plt.figure(figsize=(20,8),dpi=80)

plt.plot(x,y)

#调整x轴的刻度
_xtick_labels = ["10点{}分".format(i) for i in range(60)]
_xtick_labels += ["11点{}分".format(i) for i in range(60)]
#取步长，数字和字符串一一对应，数据的长度一样
plt.xticks(list(x)[::3],_xtick_labels[::3],rotation=45,fontproperties=my_font) #rotaion旋转的度数

#添加描述信息
plt.xlabel("时间",fontproperties=my_font)
plt.ylabel("温度 单位(℃)",fontproperties=my_font)
plt.title("10点到12点每分钟的气温变化情况",fontproperties=my_font)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the chart
years = [2023, 2024, 2029]
north_america = [18, 20, 40]
europe = [8, 10, 20]
asia_pacific = [5, 7, 15]
rest_of_world = [2, 3, 6]

# Calculate the cumulative values for stacking
cumulative_europe = np.array(north_america) + np.array(europe)
cumulative_asia_pacific = cumulative_europe + np.array(asia_pacific)
cumulative_rest_of_world = cumulative_asia_pacific + np.array(rest_of_world)

# Set up the figure and axis with a dark theme
plt.figure(figsize=(12, 6), facecolor='#0D1B2A')  # Background color for the figure
ax = plt.gca()
ax.set_facecolor('#0D1B2A')  # Background color for the axis

# Plot the stacked bar chart with adapted colors
plt.bar(years, north_america, color='royalblue', edgecolor='black', label='North America')
plt.bar(years, europe, bottom=north_america, color='mediumpurple', edgecolor='black', label='Europe')
plt.bar(years, asia_pacific, bottom=cumulative_europe, color='lightseagreen', edgecolor='black', label='Asia Pacific')
plt.bar(years, rest_of_world, bottom=cumulative_asia_pacific, color='greenyellow', edgecolor='black', label='Rest of the World (RoW)')

# Add CAGR text
plt.text(2023.7, 70, "CAGR of\n34.2%", fontsize=35, color='#00FFFF', weight='bold')

# Highlight the 2029 value and draw an arrow
plt.text(2029, 96.3, "96.3", fontsize=10, weight='bold', va='bottom', ha='center', color='white')
plt.annotate('', xy=(2029, 96.3), xytext=(2024.5, 40),
             arrowprops=dict(arrowstyle="->", color='#00FFFF', lw=1.5))

# Add title and labels with light-colored fonts
plt.title('Augmented and Virtual Reality Market\nGlobal Forecast to 2029 (USD BN)', fontsize=16, weight='bold', color='#00FFFF')
plt.xlabel('Year', fontsize=12, color='white')
plt.ylabel('Market Value (USD BN)', fontsize=12, color='white')
plt.xticks(years, fontsize=10, color='white')
plt.yticks(range(0, 101, 20), fontsize=10, color='white')
plt.ylim(0, 100)

# Add legend with light-colored fonts
plt.legend(title='Regions', fontsize=10, loc='upper left', title_fontsize=10, facecolor='#0D1B2A', edgecolor='white', labelcolor='white')

# Add gridlines for better readability with lighter lines
plt.grid(axis='y', linestyle='--', alpha=0.4, color='white')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 3, 4)
y = np.linspace(0, 2, 3)

X, Y = np.meshgrid(x, y)

plt.scatter(X, Y, color='red')  # 在网格点上绘制红点
plt.title("2D Meshgrid")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```
